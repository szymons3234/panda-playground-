### Loading another useful libraries

```python
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,FunctionTransformer,MinMaxScaler
from sklearn.impute import SimpleImputer
```

### Making pipeline for categorical features

```python
categorical_features = ["sex"]
categorical_transformer = Pipeline(steps=[
    ('inpute_mode',SimpleImputer(strategy="mode"),
    ('ohe',OneHotEncoder(handle_unknown="ignore",sparse_output=False,drop='first')))
])
```

### Making pipeline for numerical features

```python
numerical_features = ['age','bmi','bp','s1','s2','s3','s4','s5','s6']
numerical_transformer = Pipeline(steps=[
    ('inpute_mean',SimpleImputer(strategy='mean'),
    ('scaler',MinMaxScaler()))
])
```

### Making column transformer

```python
col_transf = ColumnTransformer(
    transformers=[
        ('cat_transf',categorical_transformer,categorical_features),
        ('num_transf',numerical_transformer,numerical_features)],
    remainder="passthrough",
    verbose_feature_names_out=True)
```
